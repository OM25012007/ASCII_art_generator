```python
!pip install pillow

from PIL import Image
from IPython.display import HTML, display


ASCII_CHARS = ["@", "#", "S", "%", "?", "*", "+", ";", ":", ",", "."]

def resize_image(image, new_width=60):
    width, height = image.size
    aspect_ratio = height / width
    new_height = int(aspect_ratio * new_width * 0.55)
    return image.resize((new_width, new_height))

def image_to_colored_ascii(image_path, width=60, save=True, font_size=10, line_height=10):
    try:
        image = Image.open(image_path)
    except:
        print(" Unable to open image file.")
        return

    
    image = resize_image(image, width)

    
    image_rgb = image.convert("RGB")
    pixels = image_rgb.getdata()

    
    gray_image = image.convert("L")
    gray_pixels = gray_image.getdata()

    ascii_html = f"<pre style='font: {font_size}px/{line_height}px Courier New, monospace;'>"

    for i in range(0, len(gray_pixels), width):
        line = ""
        for j in range(width):
            gray_val = gray_pixels[i + j]
            char = ASCII_CHARS[gray_val // 25]

            r, g, b = pixels[i + j]
            line += f"<span style='color: rgb({r},{g},{b})'>{char}</span>"
        ascii_html += line + "\n"

    ascii_html += "</pre>"

    
    display(HTML(ascii_html))

    if save:
        with open("ascii_art_colored.html", "w") as f:
            f.write(ascii_html)
        print(" Colored ASCII art saved as ascii_art_colored.html (check output files tab)")


image_to_colored_ascii("/kaggle/input/dog-image-4/dog.jpeg",
                       width=60, font_size=10, line_height=10)
```
